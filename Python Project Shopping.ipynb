```python
 Social_Media = ['Facebook','Instagram','Snapchat', 'Twitter', 'Other']
Counts = [548, 916, 86, 179, 947]

plt.figure(1)
x = range(5)
plt.xticks(x, Social_Media)
sns.barplot(x = Social_Media, y = Counts)
plt.title("Mobile Shopping Across the Globe")
plt.xlabel("Social Media")
plt.ylabel("Votes")
```

```python
y = np.array([548, 916, 86, 179, 947])
mylabels = ["Facebook", "Instagram", "Snapchat", "Twitter", "Other"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Social Media:")
plt.legend(loc='center')
plt.title("Global Social Media Shopping Platforms")
plt.show() 
```
